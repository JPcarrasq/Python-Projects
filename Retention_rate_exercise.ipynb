```python
active_client = 0
paused_client = 0

for index, row in df.iterrows():
    if row['Cancelación suscripción'] == 0.0 or row['Cancelación suscripción'] == 28.0:  # Ajuste aquí
        paused_client += 1
    elif row['Cancelación suscripción'] == " Cancelación suscripción":
        paused_client += 1
    else:
        active_client += 1

real_client_value = paused_client

mean = real_client_value / (active_client + real_client_value)
print(mean)
print(active_client)
print(real_client_value)
```

```python
df["Llamadas caídas"].value_counts().sort_values(ascending=False)
df["Llamas por quejas al mes"].value_counts().sort_values(ascending=False)

print(df["Llamadas caídas"].value_counts())
print(df["Llamas por quejas al mes"].value_counts())
```

```python
filtered_df = df[(df['Llamadas caídas'] >= 5) & (df['Llamas por quejas al mes'] >= 3)]

# Paso 2: Calcular la probabilidad
total_clientes = len(df)
clientes_cumplen_condiciones = len(filtered_df)
probabilidad = clientes_cumplen_condiciones / total_clientes

# Imprimir la probabilidad
print("Probabilidad de que un cliente seleccionado al azar cumpla las condiciones: {:.2f}".format(probabilidad))
```
